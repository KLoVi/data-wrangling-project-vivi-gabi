```python
df = df.loc[~(df['Columna1'] == 2)]

print("DataFrame después de eliminar filas:")
print(df)
```

```python
# Encontrar índices de filas a eliminar
indices_a_eliminar = df[df['Columna1'] == 2].index

# Eliminar esas filas del DataFrame
df.drop(indices_a_eliminar, inplace=True)

print("DataFrame después de eliminar filas:")
print(df)
```

```python
import pandas as pd

# Ejemplo de DataFrame
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Subcategory': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
    'Value': [10, 15, 10, 20, 30, 25]
}

df = pd.DataFrame(data)

# Crear la tabla pivot inicial con sumas
pivot = pd.pivot_table(df, 
                       values='Value', 
                       index='Category', 
                       columns='Subcategory', 
                       aggfunc='sum', 
                       margins=True, 
                       margins_name='Total')

# Calcular los porcentajes basados en los totales por fila
pivot_percentage = pivot.div(pivot.loc[:, 'Total'], axis=0) * 100

print(pivot)
print(pivot_percentage)
```
