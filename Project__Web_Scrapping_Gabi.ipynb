```python
import requests
from bs4 import BeautifulSoup
import json

# URL of the webpage
url = 'https://news.gallup.com/poll/394025/world-unhappier-stressed-ever.aspx'

# Send a GET request to fetch the HTML content
response = requests.get(url)
html_content = response.content

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find the <path> element with the specific 'aria-datavaluearray' attribute
data_element = soup.find('path', {'aria-datavaluearray': True})

# Check if the element is found
if data_element:
    # Extract the value of the 'aria-datavaluearray' attribute
    data_array_str = data_element['aria-datavaluearray']

    # Parse the JSON-like string into a Python list
    data_array = json.loads(f'[{data_array_str}]')

    # Convert the list into a dictionary
    data_dict = {year: value for year, value in data_array}

    # Print the resulting dictionary
    print(data_dict)
else:
    print("Data element not found.")
```
