## "World Unhappier, More Stressed Out Than Ever"

Source: Gallup, 01.07.24

https://news.gallup.com/poll/394025/world-unhappier-stressed-ever.aspx

#### Negative Experience Index

"Gallup asked adults in 122 countries and areas in 2021 if they had five different negative experiences on the day before the survey -- and compiled the results into an index. Higher scores on the Negative Experience Index indicate that more of a population is experiencing these emotions"

Worry, 
Stress, 
Physical pain, 
Sadness, 
Anger

#### Positive Experience Index

"The Positive Experience Index is based on people's responses to five questions about positive experiences they had the day before the survey. Higher scores indicate that more of the population reported experiencing these emotions.

Well-rested, 
Enjoyment, 
Smiled or laughed 
Treated with respect 
Learned or did something interesting the day before the interview

```python


# Parse the webpage content with BeautifulSoup
soup_idx = BeautifulSoup(response.content, 'html.parser')

# Find the element with the specific attribute
data_div = soup_idx.find('div', {'aria-datavaluearray': True})

# Extract the value of the 'aria-datavaluearray' attribute
data_values = data_div['aria-datavaluearray']

# Print the extracted data values
print(data_values)

# If you need to process the data, you can convert it from string to list
import ast
data_list = ast.literal_eval(data_values)

# Print the processed data list
print(data_list)
```
